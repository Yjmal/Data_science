```python
parsedData = data.replace('\t','\n').split('\n')
```

```python
labels = parsedData[0::2]
text = parsedData[1::2]
print(len(labels))
print(len(text))
```

```python
import pandas as pd 
df = pd.DataFrame({
    'label' : labels[:-1],
    'text'  : text
})
df.head()
```

### remove punctuation 

```python
import string 
def remove_punct(text):
    text_clean = "".join([char for char in text if char not in string.punctuation])
    return text_clean 
```

```python
df['text_clean'] = df['text'].apply(lambda x : remove_punct(x))
df.head()
```

### tokenization

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts(df['text_clean'])
df['text_tokenized'] = tokenizer.texts_to_sequences(df['text_clean'])
df.head()
```

```python
###manual method using re 
import re 
def tokenize(text):
    tokens = re.split('\W+',text)
    return tokens 
df['text_tokenized_re'] = df['text_clean'].apply(lambda x: tokenize(x.lower()))
df.head()
```

### stopwords

```python
import nltk 
stopword  = nltk.corpus.stopwords.words('english')
df['text_nostop'] = df['text_tokenized_re'].apply(lambda x : [word for word in x if word not in stopword])
df.head()
```

### lemmatization

```python
ps = nltk.PorterStemmer()
```

```python
df['text_stemmed'] = df['text_nostop'].apply(lambda x : [ps.stem(word) for word in x ])
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vec = CountVectorizer()
```

```python
df['text_final'] = df['text_stemmed'].apply(lambda x : " ".join(x))
df.head()
```

```python
X_count = count_vec.fit_transform(df['text_final'])
print(X_count)
X_count.toarray()
```

```python
def clean_text(text) :
    text_clean = "".join([char for char in text if char not in string.punctuation])
    tokens = re.split('\W+',text_clean)
    text_stemmed = [ps.stem(word) for word in tokens if word not in stopword]
    final_text = " ".join(text_stemmed)
    return text_stemmed 
```

```python
count_vec2 = CountVectorizer(analyzer = clean_text)
X_count2 = count_vec2.fit_transform(df['text'])
print(X_count2)
```

```python
X = X_count2.toarray()
y = df.iloc[:,0].values 
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
y = lb.fit_transform(y)
y
```

```python
from sklearn.model_selection import train_test_split 
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 0)
```

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score 
print(classification_report(y_test,y_pred))
accuracy_score(y_pred,y_test)
```
